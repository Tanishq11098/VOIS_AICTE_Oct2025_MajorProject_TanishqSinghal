```python
# Replace 'your_file_name.xlsx' with the actual name of your file
file_path = r'/content/Netflix Dataset 1.xlsx'
try:
    df = pd.read_excel(file_path)
    print(f"Successfully loaded {file_path}")
    display(df.head())
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred while reading the file: {e}")
```

# Task
Analyze the Netflix dataset located at "D:\Finance\Internship\VOIS\netflix" to show a bar chart comparing the count of Movies vs TV Shows for each year from 2008 to 2021.

## Process data

### Subtask:
Convert the 'Release_Date' column to datetime objects and extract the year.


**Reasoning**:
Convert the 'Release_Date' column to datetime objects and extract the year into a new column.



```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
df['Release_Year'] = df['Release_Date'].dt.year
display(df.head())
```

**Reasoning**:
The previous attempt to convert 'Release_Date' to datetime failed due to inconsistent format. Using `errors='coerce'` will turn unparseable dates into NaT.



```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['Release_Year'] = df['Release_Date'].dt.year
display(df.head())
```

## Filter data

### Subtask:
Filter the DataFrame to include only releases between the years 2008 and 2021 (inclusive).


**Reasoning**:
Filter the DataFrame to include releases between 2008 and 2021 and display the head of the filtered DataFrame.



```python
df_filtered = df[(df['Release_Year'] >= 2008) & (df['Release_Year'] <= 2021)]
display(df_filtered.head())
```

## Group and count

### Subtask:
Group the filtered data by release year and 'Category', then count the number of titles in each group.


**Reasoning**:
Group the filtered data by release year and 'Category', count the titles, reset the index, and rename the count column to prepare for visualization.



```python
yearly_counts = df_filtered.groupby(['Release_Year', 'Category']).size().reset_index(name='Count')
display(yearly_counts.head())
```

## Visualize data

### Subtask:
Create a grouped bar chart showing the count of Movies and TV Shows for each year from 2008 to 2021.


**Reasoning**:
Generate a grouped bar chart to visualize the yearly counts of movies and TV shows, incorporating the specified labels, title, and x-axis rotation for clarity.



```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Release_Year', y='Count', hue='Category', data=yearly_counts)
plt.title("Number of Movies and TV Shows Released Per Year (2008-2021)")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Summary:

### Data Analysis Key Findings

* The analysis successfully filtered the Netflix dataset to include releases between 2008 and 2021.
* The count of Movies and TV Shows released each year within the specified period was determined.
* A grouped bar chart was generated to visualize the yearly counts of Movies versus TV Shows.

### Insights or Next Steps

* The generated visualization provides a clear comparison of the volume of Movies and TV Shows added to Netflix annually between 2008 and 2021.
* Further analysis could explore the trend of Movie vs TV Show releases over time and investigate factors that might influence these trends.

# Task
Visualize a histogram of content duration to analyze runtime patterns for Movies and TV Shows using the data from "D:\Finance\Internship\VOIS\netflix".

## Process data

### Subtask:
Clean and convert the 'Duration' column to numerical format (minutes for movies, seasons for TV shows). Handle missing or invalid values.


**Reasoning**:
Create a copy of the DataFrame, define a function to convert 'Duration' based on 'Category' to numerical format, apply the function to create a new column, and display the head of the processed DataFrame.



```python
df_duration = df.copy()

def convert_duration(row):
    if pd.isna(row['Duration']):
        return np.nan
    if row['Category'] == 'Movie':
        try:
            return int(row['Duration'].split(' ')[0])
        except ValueError:
            return np.nan
    elif row['Category'] == 'TV Show':
        try:
            return int(row['Duration'].split(' ')[0])
        except ValueError:
            return np.nan
    return np.nan

df_duration['Numerical_Duration'] = df_duration.apply(convert_duration, axis=1)
display(df_duration.head())
```

## Filter data

### Subtask:
Separate the DataFrame into two based on the 'Category' column: one for Movies and one for TV Shows.


**Reasoning**:
Separate the DataFrame into movies and TV shows based on the 'Category' column and display the head of both resulting DataFrames.



```python
movies_df = df_duration[df_duration['Category'] == 'Movie'].copy()
tv_shows_df = df_duration[df_duration['Category'] == 'TV Show'].copy()

print("Movies DataFrame Head:")
display(movies_df.head())

print("\nTV Shows DataFrame Head:")
display(tv_shows_df.head())
```

## Visualize data

### Subtask:
Create separate histograms for the duration of Movies and the duration of TV Shows.


**Reasoning**:
Generate two histograms side by side to visualize the duration distribution for movies and TV shows, with appropriate labels and titles.



```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Histogram for Movie Durations
sns.histplot(ax=axes[0], data=movies_df, x='Numerical_Duration', bins=20, kde=True)
axes[0].set_title("Distribution of Movie Durations")
axes[0].set_xlabel("Duration (minutes)")
axes[0].set_ylabel("Frequency")

# Histogram for TV Show Durations
sns.histplot(ax=axes[1], data=tv_shows_df, x='Numerical_Duration', bins=20, kde=False)
axes[1].set_title("Distribution of TV Show Seasons")
axes[1].set_xlabel("Number of Seasons")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   The distribution of movie durations is right-skewed, with a peak frequency between approximately 50 and 100 minutes.
*   The distribution of TV show durations (measured in seasons) shows a high frequency of shows with a small number of seasons, with a sharp decline as the number of seasons increases.

### Insights or Next Steps

*   Netflix primarily features movies with runtimes under 100 minutes and TV shows with fewer seasons. This suggests a focus on content that is relatively easy to consume.
*   Further analysis could investigate if there is a correlation between content duration/seasons and popularity or viewer engagement.
